```python
from sklearn.datasets import load_iris

df = load_iris()
```

```python
x = df.data
y = df.data
```

```python
x
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
model.predict([[1,1,1,1]])
```

```python
import dtreeviz
model_d =dtreeviz.model(model, x,y, feature_name=df.feature_names, target_name='df', class_name=df.target_names)
v =model_d.view()
v.show()
```
