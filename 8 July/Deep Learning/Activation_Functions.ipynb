```python
# for output layer

def Sigmoid_b(x):
    return(1/(1+np.exp(-x)))

x = np.linspace(-5,5)
plt.plot(x, Sigmoid_b(x))
plt.title('Activation - binary sigmoid function')
plt.show()
```

```python
def TanH(x):
    return np.tanh(x)

x = np.linspace(-5,5)
plt.plot(x, TanH(x))
plt.title('Activation TanH function')
plt.show()
```

```python
#for hidden layer (relu function)
def RectifiedLinearUnit(x):
    list=[]
    for i in x:
        if i>0:
            list.append(i)
        else:
            list.append(0)
    return list

x = np.linspace(-5,5)
plt.plot(x, RectifiedLinearUnit(x))
plt.title('Activation - RELU (RectifiedLinearUnit) function')
plt.show()
```

```python
#for hidden layer (relu function) - won't work for negative values
def RectifiedLinearUnit(x):
    list=[]
    for i in x:
        if i>0:
            list.append(i)
        else:
            list.append(0)
    return list

x = np.linspace(-1000,0)
plt.plot(x, RectifiedLinearUnit(x))
plt.title('Activation - RELU (RectifiedLinearUnit) function')
plt.show()
```

```python
def LeakyreLu(x):
    list =[]
    for i in x:
        if i>0:
            list.append(i)
        else:
            list.append(0.01*i)
    return list

x = np.linspace(-5,5)
plt.plot(x, LeakyreLu(x))
plt.title('Activation- LeakyRELU function')
plt.show()
```

```python
# for negative values

def LeakyreLu(x):
    list =[]
    for i in x:
        if i>0:
            list.append(i)
        else:
            list.append(0.01*i)
    return list

x = np.linspace(-1000,5)
plt.plot(x, LeakyreLu(x))
plt.title('Activation- LeakyRELU function')
plt.show()
```

```python
def softmax(x):
    return (np.exp(x)/np.sum(np.exp(x)))


x = np.linspace(-5,5)
plt.plot(x, softmax(x))
plt.title('Activation- softmax function')
plt.show()
```

```python
inputs = np.array([4,7,9,2])
weights = np.array([0.2,0.8,0.5,1.0])
bias = 2

# or :
# inputs = [1,2,3,2.5]
# weights = [0.2,0.8,-0.5,1.0]
# bias = 2

output = (inputs[0]*weights[0]+inputs[1]*weights[1]+inputs[2]*weights[2]+inputs[3]*weights[3] + bias)/4
print(output)
```
